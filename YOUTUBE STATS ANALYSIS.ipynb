category and subscribers analysis

```python
cat_subs = df.groupby('category')['subscribers'].mean().sort_values(ascending = False)
cat_subs
```

```python
plt.figure(figsize = (15, 6))
cols = ['lightgreen', 'black', 'pink', 'skyblue', 'navy', 'darkgreen', 'orange', 'gray', 'yellow', 'red', 'brown']
cat_subs.plot(kind = 'bar', edgecolor = 'black', color = cols)
plt.xticks(rotation = 45)
plt.title('category analysis', fontsize = 11)
plt.xlabel('\n\ncategory', fontsize = 9.5)
plt.ylabel('average number of subs', fontsize = 9.5)
plt.show()
```

channel analysis

```python
channel_types = df['channel_type'].value_counts()
channel_types
```

```python
plt.figure(figsize = (13, 6))
channel_types.head(5).plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('top 5 channel distribution\n')
plt.ylabel('')
plt.show()
```

subscriber and views analysis

```python
plt.figure(figsize = (12, 5))
#sns.scatterplot(x = 'subscribers', y = 'video views', data = df, color = 'red', marker = 'o')
sns.regplot(x='subscribers', y='video views', data=df, 
            scatter_kws={'color': 'pink', 's': 40},   line_kws={'color': 'deeppink', 'linewidth': 2})
plt.title('subscriber and views', fontsize = 11)
plt.xlabel('subscribers', fontsize = 9.5)
plt.ylabel('views', fontsize = 9.5)
plt.show()
```

subscribers by country

```python
subs_country = df.groupby('country')['subscribers'].sum().sort_values(ascending = False)
subs_country.head(10)
```

visualizing subs per country

```python
cols = ['lightgreen', 'black', 'pink', 'skyblue', 'navy', 'darkgreen', 'orange', 'gray', 'yellow', 'red', 'brown']
subs_country.head(10).plot(kind = 'barh', color = cols)
plt.title('country and total subs', fontsize = 11)
plt.xlabel('subs (hundreds of millions)', fontsize = 9.5)
plt.ylabel('country', fontsize = 9.5)
plt.show()
```

last 30 day subscriber analysis

```python
subs_30 = df.groupby('country')['subscribers_for_last_30_days'].sum().sort_values(ascending = False)
subs_30.head(10)
```

channel type and uploads analysis

```python
channel_uploads = df.groupby('channel_type')['uploads'].sum().sort_values(ascending = False)
channel_uploads.head(10)
```

visualizing channel type and uploads analysis

```python
plt.figure(figsize = (13, 6))
channel_uploads.head(5).plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('top categories with the most uploads\n', fontsize = 11)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

channel type and views analysis

```python
channel_analysis = df.groupby('channel_type')['video views'].sum().sort_values(ascending = False)
channel_analysis.head(10)
```

```python
plt.figure(figsize = (12, 5))
cols = ['lightgreen', 'pink', 'skyblue', 'orange', 'gray', 'yellow', 'red', 'brown']
channel_analysis.head(5).plot(kind = 'bar', color = cols, edgecolor = 'black')
plt.legend(loc = 'upper right')
plt.title('channel type and views analysis', fontsize = 11)
plt.xticks(rotation = 45)
plt.xlabel('channel type', fontsize = 9.5)
plt.ylabel('total views (in billions)', fontsize = 9.5)
plt.show()
```

channel type and views over the last 30 days **analyzed**

```python
plt.figure(figsize = (20, 5))
sns.violinplot(x = 'channel_type', y = 'video_views_for_the_last_30_days', data = df, palette = 'Set2', hue = 'channel_type')
plt.title('\nchannel type and views (last 30 days)\n', fontsize = 11)
plt.xlabel('\nchannel type', fontsize = 9.5)
plt.ylabel('views (last 30 days)', fontsize = 9.5)
plt.legend().remove()
plt.show()
```

channel type and highest monthly earnings

```python
plt.figure(figsize = (20, 5))
sns.boxplot(x = 'channel_type', y = 'highest_monthly_earnings', data = df, palette = 'Set2', hue = 'channel_type', showfliers = False)
plt.title('\nchannel type and highest monthly earnings', fontsize = 11)
plt.xlabel('\nchannel type', fontsize = 9.5)
plt.ylabel('highest monthly earnings', fontsize = 9.5)
plt.legend().remove()
plt.show()
```

video views and highest monthly earnings analysis

```python
plt.figure(figsize = (12, 5))
sns.regplot(x='video views', y='highest_monthly_earnings', data=df, 
            scatter_kws={'color': 'skyblue', 's': 40},   line_kws={'color': 'navy', 'linewidth': 2})
plt.title('video views and highest monthly earnigns', fontsize = 11)
plt.xlabel('vews', fontsize = 9.5)
plt.ylabel('highest monthly earnings', fontsize = 9.5)
plt.show()
```

categories and views over the years

```python
plt.figure(figsize = (15, 4))
sns.lineplot(x = 'created_year', y = 'video views', data = df, label = 'views', color = 'deeppink', marker = 'o')
plt.title('\nchannel creation year and views\n', fontsize = 11)
plt.xlabel('\nyear', fontsize = 9.5)
plt.ylabel('views', fontsize = 9.5)
plt.xticks(rotation = 45)
plt.show()
```

finding out which era of youtubers made the **most** money

```python
youtube_years = df.groupby(['created_year', 'youtuber'])['highest_yearly_earnings'].mean().sort_values(ascending = False).head(10)
youtube_years
```

```python
cols = ['lightgreen', 'pink','red', 'skyblue', 'orange', 'gray', 'yellow', 'brown']
plt.figure(figsize = (22, 5))
youtube_years.plot(kind = 'bar', stacked = True, color = cols)
plt.title('\nyoutube channels and highest yearly earnings\n', fontsize = 11)
plt.xlabel('\n\nchannel and year of creation', fontsize = 9.5)
plt.xticks(rotation = 25)
plt.ylabel('earnings (in hundreds of millions)', fontsize = 9.5)
plt.show()
```

```python
youtube_months = df.groupby(['created_year', 'youtuber'])['highest_monthly_earnings'].mean().sort_values(ascending = False).head(10)
youtube_months
```

```python
cols = ['lightgreen', 'pink','red', 'skyblue', 'orange', 'gray', 'yellow', 'brown']
plt.figure(figsize = (22, 5))
youtube_months.plot(kind = 'bar', stacked = True, color = cols)
plt.title('\nyoutube channels and highest monthly earnings\n', fontsize = 11)
plt.xlabel('\n\nchannel and year of creation', fontsize = 9.5)
plt.xticks(rotation = 25)
plt.ylabel('earnings (in hundreds of millions)', fontsize = 9.5)
plt.show()
```

proportion of highest monthly earners by creation year

```python
year_earners = df.groupby('created_year')['highest_monthly_earnings'].mean().sort_values(ascending = False).head(10)
year_earners
```

```python
cols =  ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
plt.figure(figsize = (8, 5))
year_earners.head(5).plot(kind = 'pie', autopct = '%1.1f%%', colors = cols)
plt.title('distribution of years that produced the highest monthly earners', fontsize = 11)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

###### _over 60% of the highest earning accounts were created in 2020 or later, fascinating stat!_
